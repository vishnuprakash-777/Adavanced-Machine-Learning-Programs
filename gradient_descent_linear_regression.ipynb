```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(X,y,learning_rate,epochs):
  m=0
  b=0
  n=len(X)
  costs = []
  for epoch in range(epochs):
    y_pred=(m*X)+b
    error=y_pred - y
    m -= learning_rate * (1/n) * np.sum(X * error)
    b -= learning_rate * (1/n) * np.sum(error)

    if epoch%100 == 0:
      cost = (1/(2*n)) * np.sum(error**2)
      costs.append(cost)
      print(f'Epochs = {epoch} , Cost = {cost}')

  plt.plot(range(0,epochs,100),costs,marker='o')
  plt.xlabel('Epochs')
  plt.ylabel('Costs')
  plt.title('Epochs vs Costs')
  plt.show()

  return m,b
```

```python
np.random.seed(42)
X=np.random.rand(100)
y=2 + 3*X + np.random.randn(100) * 0.1
learning_rate=0.01
epochs=1000
m,b=gradient_descent(X,y,learning_rate,epochs)
print(f'Slope = {m}')
print(f'Intercept = {b}')
```
